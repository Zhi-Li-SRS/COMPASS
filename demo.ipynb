Plot the reference spectra


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_ref = pd.read_csv('Raman_dataset/ten_molecules_dataset/train_reference.csv')
wavenumbers = df_ref.columns[1:].astype(float).values

plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 12
plt.rcParams['font.weight'] = 'bold'
plt.rcParams['figure.dpi'] = 100  # Lower DPI to reduce image size

plt.figure(figsize=(10, 15))
plt.style.use('seaborn-v0_8-deep')

cmap = plt.cm.tab10
colors = [cmap(i) for i in range(len(df_ref))]

for i, row in enumerate(df_ref.iterrows()):
    name = row[1]['name']
    spectrum = row[1].values[1:].astype(float)  # Skip the name column

    offset = i * 0.5
    
    plt.plot(wavenumbers, spectrum + offset, label=name, color=colors[i % len(colors)], linewidth=1.5)

plt.xlabel('Wavenumber (cm$^{-1}$)', fontsize=12, fontweight='bold')
plt.ylabel('Normalized Intensity (a.u)', fontsize=12, fontweight='bold')
# Make the legend only two rows and on the top of the figure
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.05), ncol=4, fontsize=10)
# Remove the top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.show()
```
